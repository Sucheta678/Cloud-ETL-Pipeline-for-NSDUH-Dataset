```python
from pyspark.sql.functions import col
from pyspark.ml.feature import Imputer

# Step 1: Select relevant columns
selected_columns = ['herever','herage','heryfu','herrec','heryrtot','her30use','herflag','heryr','hermon',
                    'herlottm','hergtovr','herlimit','herkplmt','herndmor','herlsefx','hercutdn','hercutev',
                    'herfmfpb','herfmctd','herserpb','herphlpb','herphctd','heraglst','herylu','hermlu',
                    'herydays','hermdays',"suicthnk", "suicplan", "suictry","k6scmon", "k6scyr", "k6scmax",
                    "spdyr", "mhsuithk", "mhsuipln", "mhsuitry"]

df_selected = df_teenager.select(*selected_columns)

# Step 2: Cast string columns to double
cols_to_cast = ['herydays', 'hermdays', 'k6scyr']
for col_name in cols_to_cast:
    df_selected = df_selected.withColumn(col_name, col(col_name).cast("double"))

# Step 3: Apply Spark Imputer
imputer = Imputer(
    inputCols=cols_to_cast,
    outputCols=[f"{c}_imputed" for c in cols_to_cast]
).setStrategy("mean")  # or "median"

df_imputed = imputer.fit(df_selected).transform(df_selected)

# Step 4: Drop old columns and rename
for c in cols_to_cast:
    df_imputed = df_imputed.drop(c).withColumnRenamed(f"{c}_imputed", c)

# df_imputed is now your final DataFrame with imputed numeric columns
```

```python
from pyspark.sql.functions import col, count

def missing_percent(df_imputed,hermdays):
    """Calculate percentage of missing values for a column"""
    total = df_imputed.count()
    missing = df_imputed.filter(
        col(hermdays).isNull() | 
        (col(hermdays) == '') |
        (col(hermdays) == 'NA') |
        (col(hermdays) == 'N/A')
    ).count()
    percent = (missing / total) * 100
    return percent

# Usage
percent_missing = missing_percent(df_imputed, 'hermdays')
print(f"Missing: {percent_missing:.2f}%")
```

```python
from pyspark.sql.functions import col

# List of numeric columns
numeric_cols = [c[0] for c in df_cleaned.dtypes if c[1] in ('int', 'double', 'float')]

# Function to compute IQR bounds for a column
def get_iqr_bounds(df_cleaned, col_name):
    quantiles = df_cleaned.approxQuantile(col_name, [0.25, 0.75], 0.05)
    Q1 = quantiles[0]
    Q3 = quantiles[1]
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

# Show outliers for each numeric column
for col_name in numeric_cols:
    lower, upper = get_iqr_bounds(df_cleaned, col_name)
    outliers = df_cleaned.filter((col(col_name) < lower) | (col(col_name) > upper))
    print(f"Outliers in column {col_name}: {outliers.count()}")
```

```python
from pyspark.sql.types import NumericType

numeric_cols = [f.name for f in df_cleaned.schema.fields if isinstance(f.dataType, NumericType)]
print(numeric_cols)
```

```python
from pyspark.sql.functions import log1p, col

df_log = df_cleaned
for c in numeric_cols:
    df_log = df_log.withColumn(c, log1p(col(c)))
```

```python
from pyspark.sql.functions import when, col

df_final = df_log

for c in numeric_cols:
    # Calculate exact quantiles (very accurate)
    lower, upper = df_final.approxQuantile(c, [0.005, 0.995], 0.001)

    df_final = df_final.withColumn(
        c,
        when(col(c) < lower, lower)
        .when(col(c) > upper, upper)
        .otherwise(col(c))
    )
```

```python
for c in numeric_cols:
    low, high = df_final.approxQuantile(c, [0.005, 0.995], 0.001)

    count = df_final.filter((col(c) < low) | (col(c) > high)).count()
    print(f"Outliers in column {c}: {count}")
```

```python
from pyspark.sql.functions import col, when

df_result = df_final.withColumn(
    "plan_and_try",
    when((col("mhsuipln") == 1) & (col("mhsuitry") == 1), 1).otherwise(0)
)
```
